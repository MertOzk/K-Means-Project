## Building the K-Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
```

```python
kmeans.fit(df.drop('Private', axis=1))
```

Looking at the cluster center vectors

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

## Evaluation

Creating a new column called cluster, 1 for Private and 0 for Public School

```python
sns.set_style('whitegrid')
sns.lmplot(data = df, x = 'Outstate', y = 'F.Undergrad', hue = 'Private', palette = 'coolwarm',fit_reg = False)
```

```python
plt.title('K Means')
plt.scatter(data=df, x = 'Outstate', y = 'F.Undergrad',c=kmeans.labels_,cmap='rainbow')
plt.legend(['Yes'])
```

If we compare the two graphs above we can see that our model is mislabeling quite a few datapoints. Lets use metrics to see the accuracy. 

```python
def converter(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

Although the model is mislabeling some datapoints, it seems like the model is doing alright taking in consideartion that it is supposed to be an unsupervised model. However the the model can definetly be improved upon to gain higher accuracy.
